# JOINTURE DANS PANDAS

```python
data1=pd.read_csv('BTC-EUR.csv',index_col='Date',parse_dates=True) 
data1
```

```python
data2=pd.read_csv('ETH-EUR.csv',index_col='Date',parse_dates=True)
data2
```

```python
data3=pd.merge(data1,data2,on ='Date',how='inner' )#comme inner join dans sql
data3
```

```python
data3=data3.drop(['Open_BTC', 'High_BTC', 'Low_BTC', 
          'Open_ETH', 'High_ETH', 'Low_ETH'  ],axis=1)
data3
```

```python
data3[['Close_BTC','Close_ETH']].plot(subplots=True)#tracer le C_BTC et C_ETH Dans de fig avec subplots  
print(data3[['Close_BTC','Close_ETH']].corr())
```

# TIME SERIE

```python
data1=pd.read_csv('BTC-EUR.csv',index_col='Date',parse_dates=True) #remplace les index par defaut par column Date
data1
```

```python
data1.index#afficher les index  
```

# RESAMPLE

```python
data1['Close'].plot()#tracer series Close
```

```python
plt.figure(figsize=(12,8))
data1['Close'].resample('Y').plot() 
```

```python
plt.figure(figsize=(12,8))
data1['Close']['2019'].resample('M').plot()#tracer les donner de l'annee 2019 avec resample chaque column par un color
```

```python
plt.figure(figsize=(12,8))
data1['Close']['2016':'2019'].resample('6M').plot()#chaque 6 mois avec un color de 2016 a 2019
```

```python
plt.figure(figsize=(12,8))
data1['Close']['2019'].resample('M').mean().plot()#tracer par rapport mean
```

```python
plt.figure(figsize=(12,8) )
data1.loc['2019','Close'].plot(label="simple plot")
data1.loc['2019','Close'].resample('M').mean().plot(label='plot + resample M',ls=':')
data1.loc['2019','Close'].resample('W').mean().plot(label='plot + resample W',ls='--')
plt.legend()
plt.show()
```

# AGG

```python
data1.loc['2019','Close'].resample('M').agg(['mean','std','min'])#agg on calcule pour chaque mois de l'annee 2019
```

```python
m=data1.loc['2019','Close'].resample('M').agg(['mean','max','min']) 
m['mean'].plot(label='moyenne plot')
plt.fill_between(m.index,m['min'],m['max'],alpha=0.2,label='max-min')#fillbetween c-a-d tracer un graphe entre min et max
plt.legend()
```

# Rolling et EWM

```python
data1.loc['2019-09','Close'].plot(label='1')
data1.loc['2019-09','Close'].rolling(window=7).mean().plot(ls=':',label='2' )#Moving Average 
data1.loc['2019-09','Close'].rolling(window=7,center=True).mean().plot(ls=':',label='3' )
plt.legend() 
```

```python
data= pd.date_range('1/1/2022', periods = 1000000, freq ='H').to_list()

# create dataframe- date column for the date data
data = pd.DataFrame(data,columns=['date'])

#add values column to the dataframe
data['values']= list(np.random.randint( 0,100,[1000000] )) 
# display 
print(np.nanmax(data['values'].rolling(5).mean())) 
```

```python
data1.loc['2019-09','Close'].plot(label='1')
data1.loc['2019-09','Close'].rolling(window=7).mean().plot(ls=':',label='2' )
data1.loc['2019-09','Close'].rolling(window=7,center=True).mean().plot(ls=':',label='3' )
data1.loc['2019-09','Close'].ewm(alpha=0.5).mean().plot(ls=':',label='4' )#like rolling but plus spicific
# exponential weigthed function ewm() comme rolling mais plus spicifie ewm(alpha)
plt.legend() 
```

```python
for i in np.arange(0.2,1,0.2):#boucle pour tracer 4 plot des cas de alpha
    data1.loc['2019-09','Close'].ewm(alpha=i).mean().plot(label=f'{i} plot',ls=':')#on voir le graph si on change alpha
plt.legend()
plt.show()    
```

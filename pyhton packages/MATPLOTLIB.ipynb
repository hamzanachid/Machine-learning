# MATPLOTLB.PYPLOT

```python
import numpy as np
import matplotlib.pyplot as plt #package pour tracer des graphs
from mpl_toolkits.mplot3d import Axes3D# pour tracer des graphs 3d
from sklearn.datasets import load_iris#data set iris
```

# Methode normale

```python
x=np.linspace(-10,10,5)#declarer la variable x
y=x**2
plt.plot(x,y)#tracer la fonction f(x) avec x
```

```python
plt.scatter(x,y)#tracer un nuage du point ou chaque x avec f(x)
```

```python
plt.plot(x,y,c='r')#donner un color du ghrap
```

```python
plt.plot(x,y,c='r',ls='--',lw=3) #ls pour changer style et lw pour changer taille
```

```python
plt.plot(x,y,c='r',ls='--',lw=6)
```

```python
plt.figure(figsize=(12,8))# figsize pour modifier taille du graph et figure pour definir une cycle de vie du graph 
plt.plot(x,y,label='Y=X^2') 
plt.plot(x,x**3,label='Y=X^3')
plt.xlabel('axe x')# donner a l'axe x un nom
plt.ylabel('axe y')# donner a l'axe y un nom
plt.title('graph X^2 and X^3')
plt.legend()
plt.show()
plt.savefig('graphsave.png')#pour sauvgarder graph
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)#subplot pour cree une matrice des graphes(subplot(nombre de ligne,nombre de column, nombre de graph))
plt.plot(x,x**2,c='r',ls='--')
plt.subplot(2,1,2)
plt.plot(x,x**3,c='y' )
```

```python
x=np.array([i for i in range(25)])
plt.figure(figsize=(12,8))
plt.plot(x,np.cos(x),c='r',label='cos')
plt.plot(x,np.sin(x),c='y',label='sin')
plt.title('cos and sin',loc='left')
font1={'family':'serif','color':'blue','size':20}#dict qui compose de plsr element pour donner a la courbe une style
font2={'family':'serif','color':'green','size':10}
plt.xlabel('x,np.cos(x)',fontdict=font1)#donner a l'axe x un style
plt.ylabel('x,np.sin(x)',fontdict=font2)
plt.legend()
```

# Methode OOP

```python
fig,ax=plt.subplots(2,1)#methode oop pour tracer ensemble des graphes 
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
fig,ax=plt.subplots(1,2,sharey=True )#les deux graphe prend la meme axe y
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

# Graph 3D

```python
a=plt.axes(projection='3d')#declarer que on veut tracer un graph 3d
f= lambda x,y:(12*np.cos((x**2+y**2)/4))/(3+x**2+y**2)#declarer fct de deux variable
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100) 
x,y=np.meshgrid(x,y)#si on n'utilise pas meshgrid shape de z = (100,1) mais on veut (100,100)  
z=f(x,y)
print(z.shape)
a.plot_surface(x,y,z)#tracer fct de deux variable
```

# SCATTER

```python
iris=load_iris()#dataset qui contien deux tableau pour chaque data prend target
print(iris)
x=iris.data#x prend la table data 
y=iris.target#y prend la table target 
print(x)
print(y)
plt.scatter(x[:,0],x[:,1],c=y)#pour chaque target on donne un colore
```

# GRID

```python
x=np.array([i for i in range(100)])
y=x**2
plt.figure()
plt.plot(x,y)
plt.grid()#cette methode pour afficher graph avec des grid 
plt.figure()
plt.plot(x,y)
plt.grid(axis='x')#cette methodee pour afficher graph avec des grid dans l'axe x 
plt.figure()
plt.plot(x,y)
plt.grid(axis='y')#cette methodee pour afficher graph avec des grid dans l'axe y 
plt.figure()
plt.plot(x,y)
plt.grid(c='r',ls='--',lw='1')#tu peux donner un color et style et taille de votre fig 
```

# BAR.BARH

```python
plt.figure()
x=np.array([1,4,7,10])#la valeur de chaque bar
y=np.array(['A','B','C','D'])#le nom de chaque bar
plt.bar(y,x,width=0.1)#avec bar tu peux cree des bar  (width pour controler la taille du bar,The default width value is 0.8 ) 
plt.figure()
x=np.array([1,4,7,10])#la valeur de chaque bar
y=np.array(['A','B','C','D'])#le nom de chaque bar
plt.barh(y,x,height=0.2)#If you want the bars to be displayed horizontally instead of vertically, use the barh() function(use height to control taille du bar)
```

```python
x=np.array([1, 2, 3, 4, 5])
 
# making subplots
fig, ax = plt.subplots(2, 2)
 
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x*x)
ax[1, 1].plot(x, x*x*x)

plt.subplots_adjust(left=0.1,#we use subplots_adjust to give panding a 
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

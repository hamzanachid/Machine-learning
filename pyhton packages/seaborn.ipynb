# SEABORN

```python
import seaborn as sns # package pour faire visualisation des donnees
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names()) #see all the data you have  in sns
data1=sns.load_dataset('car_crashes')
data2=sns.load_dataset('tips')
data3=sns.load_dataset('flights')
data4=sns.load_dataset('iris')
data5=sns.load_dataset('titanic')
data5
```

# DISTRIBUTION PLOT

```python
sns.distplot(data1['not_distracted'])#plot distribution
```

```python
sns.distplot(data1['not_distracted'],kde=False)# kde=density
```

```python
sns.distplot(data1['not_distracted'],kde=False,bins=20)#bins = number of class
```

# JOINT PLOT

```python
sns.jointplot(x='speeding',y='alcohol',data=data1) # a joint plot
```

```python
sns.jointplot(x='speeding',y='alcohol',data=data1,kind='reg')# modify kind of jointplot(reg kde hex)
```

# KDEPLOT

```python
sns.kdeplot(data1['speeding'])
```

# BARPLOT

# Categorical plots

```python
sns.barplot(x='sex',y='total_bill',data=data2,estimator=np.cov)##by default estimator=np.mean(bar plot groupby (sex).mean)
```

# COUNTPLOT

```python
sns.countplot(x='sex',data=data2)
```

# BOXPLOT

```python
sns.boxplot(x='day',y='total_bill',data=data2)
```

```python
sns.boxplot(x='sex',y='total_bill',data=data2,hue='day')
plt.legend(loc = 0)
```

# Violinplot

```python
sns.violinplot(x='sex',y='total_bill',data=data2,hue='day')
```

```python
sns.violinplot(x='day',y='total_bill',data=data2,hue='sex',split=True)
```

# Swarmplot

```python
sns.swarmplot(x='day',y='total_bill',data=data2 ,hue='sex')
sns.violinplot(x='day',y='total_bill',data=data2)
```

```python
sns.swarmplot(x='day',y='total_bill',data=data2,hue='sex',palette='afmhot')#palette(and chose a colormaps from matplotlib site)
plt.legend(loc=10)#loc to chose the best place for the legend
```

# HEATMAP

```python
corr=data1.corr()
sns.heatmap(corr  ,cmap='Blues') 
plt.show()
```

```python
corr=data1.corr()
sns.heatmap(corr,annot=True,cmap='Blues')#annot = add the number in center of graph
plt.show()
```

```python
plt.figure(figsize=(12,8))
data3=data3.pivot_table(index='month',columns='year',values='passengers')  
sns.heatmap(data3,annot=True,linecolor='white',linewidths=1)
```

# Clustermap

```python
species=data4.pop('species')
sns.clustermap(data4) 
sns.clustermap(data3,cmap='Blues',standard_scale=1)
```

# PairGrid

```python
data41=sns.PairGrid(data4 )
data41.map_upper(plt.scatter)
data41.map_lower(sns.kdeplot)
data41.map_diag(plt.hist)
```

```python
data41=sns.PairGrid(data4,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['sepal_length','sepal_width'])
data41.map(plt.scatter)
data41.add_legend()
```
